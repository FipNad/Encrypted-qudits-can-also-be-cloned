For all $k,l,m,n \in \{0,\dots,d-1\}$,
$$\text{Tr}\Big((X_d^{\scriptscriptstyle k}Z_d^{\scriptscriptstyle l}\otimes I_d)\ket{\Phi_d}\bra{\Phi_d}(Z_d^{\scriptscriptstyle -n}X_d^{\scriptscriptstyle -m}\otimes I_d)\Big) = \delta_{k,m}\delta_{l,n}.$$

```python
import sys
import os
import numpy as np
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from QuditGates import QuditGates
from QuditCircuit import QuditCircuit
from helper_functions import *
```

```python
dimension = 4
gates = QuditGates(dimension)
```

```python
X = gates.get_X_gate()
Z = gates.get_Z_gate()
I = gates.get_I_gate()
BELL_STATE = vector_to_ket(create_maximally_entangled_state(dimension))
```

We can observe in the following snippet, that the only cases when the condition is true, is achieved only for the cases where $k_1 = k_3$ and $k_2 = k_4$. For all the other cases, the trace is equal to 0, as depicted by the Kronecker-delta function.

```python
for k in range(dimension):
    for l in range(dimension):
        for m in range(dimension):
            for n in range(dimension):
                left_op = np.kron(np.linalg.matrix_power(X, k) @ np.linalg.matrix_power(Z, l), I) @ BELL_STATE
                right_op = BELL_STATE.conj().T @ np.kron(np.linalg.matrix_power(Z, -n) @ np.linalg.matrix_power(X, -m), I)
                matrix = left_op @ right_op
                trace = np.trace(matrix)
                print(f"Result for (k={k}, m={m}), (l={l}, n={n}): {np.allclose(trace, 1)}")
        print("\n")
```
