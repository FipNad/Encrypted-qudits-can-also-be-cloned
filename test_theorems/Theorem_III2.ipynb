For any operator $O_1,O_2 \in \mathcal{M}_{\mathbb{C}^{d}\times\mathbb{C}^{d}}$,
    $$\text{Tr}_B\Big((O_1\otimes I_B)\ket{\Phi_d}_{A,B}\bra{\Phi_d}_{A,B}(O_2^{\dagger }\otimes I_B)\Big) = \frac{1}{d}O_1O_2^{\dagger}.$$

```python
import sys
import os
import numpy as np
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from QuditGates import QuditGates
from QuditCircuit import QuditCircuit
from helper_functions import *
```

```python
dimension = 10
gates = QuditGates(dimension)
```

```python
BELL_STATE = vector_to_ket(create_maximally_entangled_state(dimension))
O1 = gates.get_random_unitary()
O2 = gates.get_random_unitary()
I = gates.get_I_gate()
```

```python
left_op = np.kron(O1, I)
right_op = np.kron(O2.conj().T, I)

state_left = left_op @ BELL_STATE
state_right = BELL_STATE.conj().T @ right_op
rho = state_left @ state_right

rho_B = partial_trace_qudits(rho, [1], 2, dimension) # We trace out the second subsystem
```

```python
LHS = rho_B
RHS = (1/dimension) * O1 @ O2.conj().T


print(f"LHS is equal to RHS: {np.allclose(LHS, RHS)}")
```
