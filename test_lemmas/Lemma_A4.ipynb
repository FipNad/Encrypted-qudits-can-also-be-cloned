Let $\ket{\Phi_d}$ the generalized Bell state, and $X_d, Z_d$ the generalized Pauli operators. 
For any
$k_1,k_2, k_3, k_4 \in \{0,\dots,d-1\}$, the following identity is true
$$
\Pi_{k_1,k_2} \cdot \Pi_{k_3,k_4} = \delta_{k_1,k_3}\delta_{k_2,k_4}\Pi_{k_1,k_2},
$$
given the notation of the puter product \(\Pi_{i,j} = (X_d^{i}Z_d^{j}\otimes I_d)\ket{\Phi_d}\bra{\Phi_d}(Z_d^{-j}X_d^{-i }\otimes I_d)\).

```python
import sys
import os
import numpy as np
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from QuditGates import QuditGates
from QuditCircuit import QuditCircuit
from helper_functions import *
```

```python
dimension = 2
gates = QuditGates(dimension)
```

```python
X = gates.get_X_gate()
Z = gates.get_Z_gate()
I = gates.get_I_gate()
BELL_STATE = vector_to_ket(create_maximally_entangled_state(dimension))
```

We can observe in the following snippet, that the only cases when the condition is true, is achieved only for the cases where $k_1 = k_3$ and $k_2 = k_4$. For all the other cases, the LHS is equal to 0, as depicted by the Kronecker-delta function.

```python
for k1 in range(dimension):
    for k2 in range(dimension):
        for k3 in range(dimension):
            for k4 in range(dimension):
                op_k1_k2 = np.kron((np.linalg.matrix_power(X, k1) @ np.linalg.matrix_power(Z, k2)), I) @ BELL_STATE
                op_k3_k4 = np.kron((np.linalg.matrix_power(X, k3) @ np.linalg.matrix_power(Z, k4)), I) @ BELL_STATE

                op1 = op_k1_k2 @ op_k1_k2.conj().T
                op2 = op_k3_k4 @ op_k3_k4.conj().T


                result = np.allclose(op1 @ op2, op1) 
                result2 = np.allclose(op1 @ op2, np.zeros_like(op1))
                
                print(f"Result for (k1={k1}, k3={k3}), (k2={k2}, k4={k4}). LHS = RHS ({result}). LHS = 0 ({result2})")
        print("\n")
```
