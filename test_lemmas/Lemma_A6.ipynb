For any value of
$m\in \{0,\dots,d-1\}$, the following equality is true
$$\sum_{j=0}^{d-1}e^{-i\frac{\pi (j+m)((j+m)+d\%2)}{d}} e^{i\frac{\pi j(j+d\%2)}{d}} = d\cdot\delta_{m,0},$$
where $d\geq2$ is fixed as the qudit dimension.

```python
import sys
import os
import numpy as np
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from QuditGates import QuditGates
from QuditCircuit import QuditCircuit
from helper_functions import *
```

```python
dimension = 10
```

```python
for m in range(dimension):
    result = 0
    for j in range(dimension):
        exp1 = np.exp((-1j * np.pi * (j+m)*(j+m +dimension % 2))/dimension)
        exp2 = np.exp((1j * np.pi * (j)*(j +dimension % 2))/dimension)
        result += exp1 * exp2
    print(f"Result for (m={m}): {result:.2f}=>{np.allclose(result, dimension)}")
```
