Let $\ket{\psi}$ be a qudit state, $\ket{\Phi_d}$ be the generalized Bell state, and $X_d, Z_d$ the generalized Pauli operators.
    For any $d\geq2$, the following equality is true.
    $$
    \begin{align*}
    \frac{1}{d}\sum_{m,n=0}^{d-1}\Big(\left(X_d^{m}Z_d^{n}\right)\ket{\psi} \otimes \left(X_d^{m}Z_d^{-n} \otimes I_d\right)\ket{\Phi_d}\Big)
    =\frac{1}{\sqrt{d}}\left(\sum_{p=0}^{d-1}\ket{p}\ket{p}\right)\ket{\psi}
    \end{align*}
    $$

```python
import sys
import os
import numpy as np
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from QuditGates import QuditGates
from QuditCircuit import QuditCircuit
from helper_functions import *
```

```python
dimension = 4
gates = QuditGates(dimension)
```

```python
X = gates.get_X_gate()
Z = gates.get_Z_gate()
I = gates.get_I_gate()

BELL_STATE = vector_to_ket(create_maximally_entangled_state(dimension))
PSI = vector_to_ket(random_pure_state(dimension))

LHS = np.zeros((dimension**3, 1), dtype=np.complex128)
RHS = vector_to_ket(np.kron(BELL_STATE, PSI))
```

```python
for m in range(dimension):
    for n in range(dimension):
        psi_matrix = np.linalg.matrix_power(X, m) @ np.linalg.matrix_power(Z, n)
        bell_matrix = np.kron((np.linalg.matrix_power(X, m) @ np.linalg.matrix_power(Z, -n)), I)
        left_term = psi_matrix @ PSI
        right_term = bell_matrix @ BELL_STATE

        term = np.kron(left_term, right_term)

        LHS += (1/dimension)* term
```

```python
print("LHS:")
display_state_abs(LHS, dimension, 3)
print("\n")
print("RHS:")
display_state_abs(RHS, dimension, 3)
print("\n")
print(f"The LHS and RHS are approximately equal: {np.allclose(LHS, RHS)}")
```
