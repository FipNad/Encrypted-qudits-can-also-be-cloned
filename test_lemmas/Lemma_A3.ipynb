Let $\ket{\Phi_d}$ be the generalized Bell state, and $X_d, Z_d$ the generalized Pauli operators. 
For any 
    $k_1,k_2, k_3, k_4 \in \{0,\dots,d-1\}$, the following identity holds.
    $$
    \bra{\Phi_d}(Z_d^{-k_2}X_d^{-k_1}\otimes I_d)\cdot (X_d^{k_3}Z_d^{k_4}\otimes I_d)\ket{\Phi_d} = \delta_{k_1,k_3}\delta_{k_2,k_4}$$

```python
import sys
import os
import numpy as np
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from QuditGates import QuditGates
from QuditCircuit import QuditCircuit
from helper_functions import *
```

```python
dimension = 4
gates = QuditGates(dimension)

```

```python
X = gates.get_X_gate()
Z = gates.get_Z_gate()
I = gates.get_I_gate()
BELL_STATE = vector_to_ket(create_maximally_entangled_state(dimension))
```

```python
for k1 in range(dimension):
    for k2 in range(dimension):
        for k3 in range(dimension):
            for k4 in range(dimension):
                operator_left = np.linalg.matrix_power(Z, -k2) @ np.linalg.matrix_power(X, -k1)
                operator_right = np.linalg.matrix_power(X, k3) @ np.linalg.matrix_power(Z, k4)

                left_state = BELL_STATE.conj().T @ np.kron(operator_left, I)
                right_state = np.kron(operator_right, I) @ BELL_STATE
                result = left_state @ right_state
                print(f"Result for (k1={k1}, k3={k3}), (k2={k2}, k4={(dimension - k4)%dimension}): {result[0].item():.2f}=>{np.allclose(result[0], 1)}")
        print("\n")
    print("\n")
```
