Let $\ket{\Phi_d}$ be the generalized Bell state, and $X_d, Z_d$ the generalized Pauli operators. 
For any value of $d\geq2$, the following equality holds.
$$
\sum_{m,n=0}^{d-1}(X_d^{m}Z_d^{n}\otimes I_d)\ket{\Phi_d}\bra{\Phi_d}(Z_d^{-n}X_d^{-m}\otimes I_d) = I_d^{\otimes(2)}.
$$ 

```python
import sys
import os
import numpy as np
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from QuditGates import QuditGates
from QuditCircuit import QuditCircuit
from helper_functions import *
```

```python
dimension = 10
gates = QuditGates(dimension)
```

```python
X = gates.get_X_gate()
Z = gates.get_Z_gate()
I = gates.get_I_gate()
BELL_STATE = vector_to_ket(create_maximally_entangled_state(dimension))
```

```python
result = np.zeros((dimension**2, dimension**2), dtype=np.complex128)
for m in range(dimension):
    for n in range(dimension):
        operator = np.linalg.matrix_power(X, m) @ np.linalg.matrix_power(Z, n)
        operator = np.kron(operator, I)

        operator_left = operator @ BELL_STATE
        operator_right = operator_left.conj().T
        
        result += operator_left @ operator_right

Id = np.kron(I, I)
print(f"The LHS is equal to the RHS: {np.allclose(result,Id)}")
```
