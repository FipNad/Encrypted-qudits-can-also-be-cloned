For any operator $U\in \mathcal{M}_{\mathbb{C}^{d}\times \mathbb{C}^d}$, the following equality is true.
    $$[
    (U\otimes I_d)\ket{\Phi_d} = (I_d\otimes U^{\mathsf{T}})\ket{\Phi_d}
    ]$$

```python
import sys
import os
import numpy as np
module_path = os.path.abspath(os.path.join('..'))
if module_path not in sys.path:
    sys.path.append(module_path)

from QuditGates import QuditGates
from QuditCircuit import QuditCircuit
from helper_functions import *
```

```python
dimension = 5
gates = QuditGates(dimension)
```

```python
I = gates.get_I_gate()
U = gates.get_random_unitary()
quantum_state = create_maximally_entangled_state(dimension)

LHS = np.kron(U, I) @ quantum_state
RHS = np.kron(I, U.T) @ quantum_state

```

```python
print("LHS:", LHS)
print("\n")
print("RHS:", RHS)
print("\n")
print(f"The LHS and RHS are equal: {np.allclose(LHS, RHS)}")
```
