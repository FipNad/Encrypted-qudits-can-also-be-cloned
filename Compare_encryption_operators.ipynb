## Initializations

```python
dimension = 2
number_S_qudits = 2 # Number of S qudits in the circuit
number_qudits = 1 + 2*number_S_qudits # Total number of qudits in the circuit 

gates = QuditGates(dimension) # Initialize the QuditGtes class
```

## Compare the encryption operators

#### Construct the encryption operator as given by equation (20) in the paper.

```python
targets = [0]+[2*i+1 for i in range(number_S_qudits)]
enc_matrix = gates.create_Uencryption(number_S_qudits)
```

#### Construct the encryption operator as given by equation (5) in the [Original paper](https://journals.aps.org/prl/pdf/10.1103/y4y1-1ll6).

```python
pauli_X = gates.get_X_gate()
pauli_Z = gates.get_Z_gate()
pauli_Y = gates.get_Y_gate()
pauli_I = gates.get_I_gate()

paulis = [pauli_I, pauli_X, pauli_Y, pauli_Z]
alphas = [1, 1j, -1*(1j)**(number_S_qudits+1), 1j]

enc_matrix_2 = np.zeros((dimension**(number_S_qudits+1), dimension**(number_S_qudits+1)), dtype=complex)

for mu in range(dimension**2):
    operator = paulis[mu]
    for i in range(number_S_qudits):
        operator = np.kron(operator, paulis[mu])
    enc_matrix_2 += alphas[mu]**(-1)*operator

enc_matrix_2 = 1/(dimension) * enc_matrix_2
```

```python
print(f"The two operators are equal: {np.allclose(enc_matrix, enc_matrix_2)}")
```
